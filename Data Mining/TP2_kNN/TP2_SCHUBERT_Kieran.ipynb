The mahalanobis distance is parametrized by a $d \times d$ covariance matrix $\Sigma$, in which case the metric is defined as:

### Remarks
Effect of k:
We see that accuracy has an inverse U relationship with accuracy. Initially, increasing k will increase classification accuracy, but with k too large accuracy then goes down. Intuitively, increasing k has the effect of looking in a larger neighbourhood and averaging more values (smoothing), and there is a trade-off between smoothness and precision. The best result is obtained with the Euclidean distance with k=10, achieving a 100% accuracy. 

Effect of sigma:
The Mahalanobis computes distances of points relative to a distribution by "removing" the covariance structure (standardization). The best results are obtained with a simple average of the feature variances (diag_average_cov) and no inclusion of the cross covariances for a fixed k. Counterintuively, the most complex distance measure (full_cov) produces the worst results. One would expect higher precision when the full covariance structure is supplied. 
In this setting, the best results with the Mahalanobis distance are obtained with k=10. In large datasets where there could be a covariance structure, one should prefer the full_cov version with an intermediate k.

### Unit circle
Remember that the unit cicle is given from the equation $\,  x^2 + y^2 = 1$ and in a matrix notation is written as $\, X^TX = 1$

Coding tip: you create the x variable, then y is defined from x. This means that for each x, the corresponding y value is calculated (and thus y has the same shape as x).

````python
import numpy as np
# Open transformation.py and implement unit_circle_points and draw.
from transformation import setup_plot, draw, unit_circle_points

# get axes for the plots
fig, ax = setup_plot()

# number of points to divide the space
n = 400
# get the points of the unit circle
# the shape of X should be : (2, 2*n) = (2, 800)
X = unit_circle_points(n)

# We print out the size of X, is it equal to (2, 800) ? 
print("Shape of X :", X.shape)

# draw the unit circle
draw(ax, X, n, 'green')
````

````python
# Open transformation.py and implement apply_transformation_AX.
from transformation import apply_transformation_AX

fig, ax = setup_plot()

# redraw the unit circle
draw(ax, X, n, 'green')

# define the transformation matrix
A = np.array([[0.8, 0.3], [0.3, 0.8]])



# apply the transformation
AX = apply_transformation_AX(X, A)


# We print out the size of AX, is it equal to (2, 800) ? 
print("Shape of AX :", AX.shape)

# draw the unit circle after transformation
draw(ax, AX, n, 'blue')
````

````python
# Open transformation.py and implement apply_transformation_Z.
from transformation import apply_transformation_Z

fig, ax = setup_plot()

# redraw the unit circle
draw(ax, X, n, 'green')
# redraw the transformed circle
draw(ax, AX, n, 'blue')

# apply the transformation
Z = apply_transformation_Z(X, A)

# We print out the size of Z, is it equal to (2, 800) ? 
print("Shape of Z :", Z.shape)

# draw the unit circle after transformation with condition
draw(ax, Z, n, 'red')
````

The first linear transformation A = $\begin{bmatrix}0.8&0.3\\0.3&0.8\end{bmatrix}$, where A is a rotation matrix, can be viewed as rotating the data points which induces covariance between the two covariates (in this case positive covariance). A reduces the individual variances $\sigma_{x1}$ and $\sigma_{x2}$ by 0.8 and induces a covariance $\sigma_{x1x2}=\sigma_{x2x1}$ of 0.3.

In the second case, we want to apply a linear transformation which "undoes" the first transformation. We denote Z = AX the first linear transformation on the data which rotates the unit circle. We know that $A^{-1}A=I$, so by this logic we can reverse the transformation to recover the unit circle by applying $A^{-1}Z=A^{-1}AX=IX=X$. This second linear transformation is plotted in red and is indeed the unit circle.
